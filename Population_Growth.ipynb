```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
population_estimates = pd.read_csv("Population-EstimatesData.csv")
population_estimates.head()
```

```python
# Set new index to Indicator Name
df = population_estimates.set_index("Indicator Name")
df.head()
```

```python
reduced_df = df.loc[["Population, total"]]
reduced_df
```

```python
#Move Population, Total back into a column
reduced_df2 = reduced_df.reset_index(level='Indicator Name')
reduced_df2
```

```python
# Set new index to Country Name
reduced_df3 = reduced_df2.set_index("Country Name")
reduced_df3.head()
```

```python
# Grab the eight regions we'll be using
reduced_df4 = reduced_df3.loc[["Central Europe and the Baltics", "East Asia & Pacific", "European Union",
                                "Latin America & Caribbean", "Middle East & North Africa", "South Asia", "Sub-Saharan Africa", "United States"]]
reduced_df4
```

```python
# Create a new dataframe with all the columns we'll be using
reduced_df4 = reduced_df4[["1970","1975","1980","1985","1990","1995","2000","2005","2010","2015","2020","2025","2030","2035","2040","2045","2050"]]
reduced_df4
```

```python
#check the datatypes
reduced_df4.dtypes
```

```python
#Create New df
reduced_df5=reduced_df4.copy()
```

```python
#convert the date columns to integers
cols = ["1970","1975","1980","1985","1990","1995","2000","2005","2010","2015","2020","2025","2030","2035","2040","2045","2050"]
reduced_df5[cols] = reduced_df5[cols].applymap(np.int64)
reduced_df5
```

```python
reduced_df5.dtypes
```

```python
# Create a series for 1970-2050
series1 = reduced_df5.loc["Central Europe and the Baltics"]
series2 = reduced_df5.loc["East Asia & Pacific"]
series3 = reduced_df5.loc["European Union"]
series4 = reduced_df5.loc["Latin America & Caribbean"]
series5 = reduced_df5.loc["Middle East & North Africa"]
series6 = reduced_df5.loc["South Asia"]
series7 = reduced_df5.loc["Sub-Saharan Africa"]
series8 = reduced_df5.loc["United States"]
```

```python
# Create a list of the years that we will use as our x axis
years = [1970,1975,1980,1985,1990,1995,2000,2005,2010,2015,2020,2025,2030,2035,2040,2045,2050]
plt.figure(figsize=(10,6))
# Plot line
plt.plot(years, series1, color="blue", label="Central Europe and the Baltics")
plt.plot(years, series2, color="green", label="East Asia & Pacific")
plt.plot(years, series3, color="red", label="European Union")
plt.plot(years, series4, color="cyan", label="Latin America & Caribbean")
plt.plot(years, series5, color="magenta", label="Middle East & North Africa")
plt.plot(years, series6, color="yellow", label="South Asia")
plt.plot(years, series7, color="orange", label="Sub-Saharan Africa")
plt.plot(years, series8, color="purple", label="United States")

# Place a legend on the chart and set our other labels
plt.legend(loc="best")
plt.title('Regional Populations Through Time', fontdict={'fontweight':'bold', 'fontsize': 18})
plt.ylabel("Population, billions", fontsize=14)
legend = plt.legend(loc="best")
legend.get_frame().set_facecolor('none')
plt.gca().legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.grid(True)
#Save the plotted figures as .pngs
plt.savefig("populations_over_time.png")

# Print our chart to the screen
plt.show()
```

### Question: 
### What does the population distribution between different regions of the world look like from 1970 through projected 2050?

* East Asia and the Pacific, South Asia, and Sub Saharan Africa lead the charge in future global population growth.
* However East Asia and the Pacific does level off with just under 2.5billion people by 2030.
* The European Union and Central Europe and the Baltics see little to no population growth in the last fifty years, nor are they projected to in the next thirty.
* The U.S. also only increases its population nominally 1970-2050 never reaching 400,000,000, and Latin America and the Caribbean levels off by 2040 with about 760,000,000.
