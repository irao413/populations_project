```python
 # Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read Population-EstimatesData into pandas
pop_df=pd.read_csv('/Users/mikan/Documents/Data Science Bootcamp/populations_project/Population-EstimatesData.csv')
pop_df.head()
```

```python
# Select the necessary rows and columns
# Rows - European Union, Central Europe and the Baltics, Sub Saharan Africa, United States, Latin America and Caribbean, 
# Middle East and North Africa, East Asia and Pacific, South Asia, World
# Columns Country Name, Country Code, Indicator Name, Indicator Code, 1970 - 2050

pop_df=pop_df.rename(columns={'Country Name':'Country_Name',
                     'Country Code':'Country_Code',
                     'Indicator Name':'Indicator_Name',
                     'Indicator Code':'Indicator_Code'})
pop_df.head()
```

```python
# Select columns that are necessary 
data1970_df=pop_df.loc[:, ['Country_Name','Indicator_Name', '1970']]
data1970_df
```

```python
# Select countries / regions
names = ['European Union', 'Central Europe and the Baltics', 'Sub-Saharan Africa', 'United States',
        'Latin America & Caribbean', 'Middle East & North Africa', 'East Asia & Pacific', 'South Asia']

data1970_df2= data1970_df[data1970_df.Country_Name.isin(names)]
data1970_df2.head()
```

```python
# Select population total row from each region
data1970_df3=data1970_df2.loc[data1970_df['Indicator_Name'] == 'Population, total']
data1970_df3
```

```python
# Create the similar table for 2020

# Select columns that are necessary 
data2020_df=pop_df.loc[:, ['Country_Name','Indicator_Name', '2020']]
data2020_df

# Select necessary regions / country names
names = ['European Union', 'Central Europe and the Baltics', 'Sub-Saharan Africa', 'United States',
        'Latin America & Caribbean', 'Middle East & North Africa', 'East Asia & Pacific', 'South Asia']

data2020_df2= data2020_df[data2020_df.Country_Name.isin(names)]
data2020_df2.head()

# Select necessary indicator name
data2020_df3=data2020_df2.loc[data2020_df['Indicator_Name'] == 'Population, total']
data2020_df3
```

```python
# Create the similar table for 2050

# Select columns that are necessary 
data2050_df=pop_df.loc[:, ['Country_Name','Indicator_Name', '2050']]
data2050_df

# Select necessary regions / country names
names = ['European Union', 'Central Europe and the Baltics', 'Sub-Saharan Africa', 'United States',
        'Latin America & Caribbean', 'Middle East & North Africa', 'East Asia & Pacific', 'South Asia']

data2050_df2= data2050_df[data2050_df.Country_Name.isin(names)]
data2050_df2.head()

# Select necessary indicator name
data2050_df3=data2050_df2.loc[data2050_df['Indicator_Name'] == 'Population, total']
data2050_df3
```

```python
total_pop50 = int(data2050_df3["2050"].sum())
total_pop50 # 3465294538.0
```

```python
# Calculate total population in the selected regions in 1970, 2020, 2050

total_pop70 = int(data1970_df3["1970"].sum())
total_pop70 # 3465294538

total_pop20 = int(data2020_df3["2020"].sum())
total_pop20 # 7409282000

total_pop50 = int(data2050_df3["2050"].sum())
total_pop50  # 9293664000

# Calculate % of population for each region for each year
percentage_pop70 = data1970_df3.groupby(["Country_Name", "Indicator_Name"]).sum() 
percentage_pop70
```

```python
# Generate pie chart - 1970
pies = ["Central Europe and the Baltics", "East Asia & Pacific", "European Union", "Latin America & Caribbean", 
        "Middle East & North Africa", "South Asia", "Sub-Saharan Africa", "United States"]
pie_pop = [9.913455e+07,1.290384e+09,4.419862e+08,2.860272e+08,
           1.384731e+08,7.137113e+08,2.905262e+08,2.050520e+08]
explode = (0,0.1,0,0,0,0,0,0)

fig1, ax1 = plt.subplots()
ax1.pie(pie_pop, explode=explode, autopct='%1.1f%%', textprops=dict(color="w"),
        shadow=True, startangle=90)

ax1.legend(pies,
          title="Regions",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_title(f"1970 Population Distribution\n Total Population: {total_pop70}")

plt.savefig("/Users/mikan/Documents/Data Science Bootcamp/populations_project/Pie1970.png")

plt.show()
```
